```python
import numpy as np 
import matplotlib.pyplot as plt 
```

Defining matrix. A and vector. b by 
A =[1 2
    3 9]
b =[1
     4]

```python
A = np.array([[1, 2], [3, 9]]) 
b = np.array([1, 4]) 
```

```python
np.set_printoptions(precision=2) 
x = np.linalg.solve(A, b) 
print(x) 
```

```python
import numpy.linalg as LA 
x = LA.solve(A, b) 
print(x)
```

The inverse matrix of. A can be easily computed by 

```python
A_inv = LA.inv(A) 
print(A_inv)
```

Defining Functions



```python
def sign(x): 
    if x < 0: 
        return -1 
    elif x > 0: 
        return 1 
    else: 
        return 0 
```

**Drawing Graphs **

                    y1(t) = sin(t), y2(t) = cos(t), t ∈ [0, 2π].
    constructs an array of 400 linearly spaced values from. 0 to.2π, where. π is denoted by np.pi

```python
t = np.linspace(0, 2 * np.pi, 400) 
y1 = np.zeros(400) 
y2 = np.zeros(400) 

for k in range(400): 
    y1[k] = np.sin(t[k]) 
    y2[k] = np.cos(t[k]) 

plt.figure(figsize=(10,6)) 
plt.plot(t, y1, label='sin(t)', color='blue') 
plt.plot(t, y2, label='cos(t)', color='red') 

plt.title('Graphs of sin(t) and cos(t)') 
plt.xlabel('t') 
plt.ylabel('y') 
plt.legend(loc='lower left') 
plt.grid(True) 
plt.show() 
```
